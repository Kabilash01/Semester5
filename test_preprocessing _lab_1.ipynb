```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
text= "hi there this is a test sentence for implementing stemming and lemmatization"

tokens = word_tokenize(text.lower())

stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in tokens if word.isalpha() and word not in stop_words]

stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(word)for word in filtered_tokens]

lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("Original Text:", tokens)
print("After Stop Word Removal:", filtered_tokens)
print("After Stemming:", stemmed_tokens)
print("After Lemmatization:", lemmatized_tokens)
```

```python
nlp = spacy.load("en_core_web_sm")
text= "hi there this is a test sentence for implementing stemming and lemmatization"
doc = nlp(text)
tokens = [token.text.lower() for token in doc]
tokens_no_stop = [token.text for token in doc if not token.is_stop and token.is_alpha]
lemmetized_tokens = [token.lemma_ for token in doc if not token.is_stop and token.is_alpha]

print("Original Text:", tokens)
print("After Stop Word Removal:", tokens_no_stop)
print("After Lemmatization:", lemmetized_tokens)
```
