```python
import spacy
nlp = spacy.load("en_core_web_sm")
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")
for token in doc:
    print(f"{token.text:<12} {token.pos_:<10} {token.tag_:<6} ")
```

```python
for token in doc:
    print(f"{token.text} {token.pos_} {token.tag_} ")
```
